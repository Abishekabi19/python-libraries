# Matplotlib labels and Title

we can use the **xlabel() and ylabel()** functions to set a label for the x- and y-axis.



```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([80,85,90,95,100,105,200,205])
y=np.array([230,240,250,260,270,280,290,300])

plt.xlabel('average pulse')
plt.ylabel('calori burnage')
plt.plot(x,y)
plt.show()
```

 you can use the **title()** function to set a title for the plot.

```python
x=np.array([80,85,90,95,100,105,200,205])
y=np.array([230,240,250,260,270,280,290,300])

plt.xlabel('average pulse')
plt.ylabel('calori burnage')
plt.title('sports data')
plt.plot(x,y)
plt.show()
```

We can use the **fontdict** parameter in **xlabel(), ylabel()**, and **title()** to set font properties for the title and labels.

```python
x=np.array([80,85,90,95,100,105,200,205])
y=np.array([230,240,250,260,270,280,290,300])

font1={'family':'serif','color':'green','size':20}
font2={'family':'serif','color':'red','size':15}

plt.xlabel('average pulse',fontdict=font2)
plt.ylabel('calori burnage',fontdict=font2)
plt.title('sports data',fontdict=font1)
plt.plot(x,y)
plt.show()
```

We can use the **loc** parameter in **title()** to position the title.

Legal values are: 'left', 'right', and 'center'. Default value is 'center'.



```python
x=np.array([80,85,90,95,100,105,200,205])
y=np.array([230,240,250,260,270,280,290,300])

plt.xlabel('average pulse')
plt.ylabel('calori burnage')
plt.title('sports data',loc='right')
plt.plot(x,y)
plt.show()
```

# Matplotlib subplots

What if we need more than one graph, in a single plot.
Which means, subplots.

We can use **subplots()** function.


The **subplots()** function takes three arguments that describes the layout of the figure.

plt.subplot(1, 2, 1)::
the figure has 1 row, 2 columns, and this plot is the first plot.

plt.subplot(1, 2, 2)::
the figure has 1 row, 2 columns, and this plot is the second plot.


```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([1,2,3,4,5])
y=np.array([10,20,30,40,45])
plt.subplot(1,2,1)
plt.plot(x,y)

x=np.array([1,3,5,7,9])
y=np.array([10,5,15,25,20])
plt.subplot(1,2,2)
plt.plot(x,y)
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=np.array([10,20,30,40,45])
plt.subplot(2,2,1)
plt.plot(x,y)

x=np.array([1,3,5,7,9])
y=np.array([10,5,15,25,20])
plt.subplot(2,2,2)
plt.plot(x,y)

x=np.array([1,2,3,4,5])
y=np.array([10,20,30,40,45])
plt.subplot(2,2,3)
plt.plot(x,y)
plt.show()
```

# Matplotlib scatter

With Pyplot, you can use the **scatter()** function to draw a scatter plot.

The scatter() function plots one dot for each observation. It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis:


```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)
plt.show()
```

We can set your own color for each scatter plot with the **color or the c argument:**

Color Each Dot

We can even set a specific color for each dot by using an array of colors as value for the c argument.


```python
x = np.array([5,7,8,7,2,17])
y = np.array([99,86,87,88,111,86])
clo=np.array(['red','black','blue','green','magenta','yellow'])
plt.scatter(x,y,c=clo)
plt.show()
```

## color map

The Matplotlib module has a number of available colormaps.

A colormap is like a list of colors, where each color has a value that ranges from 0 to 100.


This colormap is called '**viridis**' and as you can see it ranges from 0, which is a purple color, and up to 100, which is a yellow color.

You can specify the colormap with the keyword argument **cmap** with the value of the colormap, in this case '**viridis**' which is one of the built-in colormaps available in Matplotlib.

In addition you have to create an array with values (from 0 to 100), one value for each of the point in the scatter plot:

You can include the colormap in the drawing by including the **plt.colorbar()** statement

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

colors=np.array([10,15,20,0,5,30,50,100,90,70,65,40,80])

plt.scatter(x,y,c=colors,cmap='viridis')
plt.colorbar()
plt.show()
```

## size

You can change the size of the dots with the s argument.

Just like colors, make sure the array for sizes has the same length as the arrays for the x- and y-axis:


```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

si=np.array([10,15,20,1,5,30,50,100,90,70,65,40,80])
plt.scatter(x,y,s=si)
plt.show()
```

## Alpha

You can adjust the transparency of the dots with the **alpha** argument. alpha range(0 to 1)

Just like colors, make sure the array for sizes has the same length as the arrays for the x- and y-axis


```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

sizes=np.array([10,15,20,1,5,30,50,100,90,70,65,40,80])
plt.scatter(x,y,s=sizes,alpha=0.3)
plt.show()
```

## Combine Color Size and Alpha

You can combine a colormap with different sizes on the dots. This is best visualized if the dots are transparent


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(50, size=(50))
y = np.random.randint(50, size=(50))

colors = np.random.randint(100, size=(50))
sizes = 10 * np.random.randint(100, size=(50))
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.colorbar()
plt.show()
```

# Matplotlib Pie Charts

With Pyplot, you can use the **pie()** function to draw pie charts.

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([30,27,40,15])
plt.pie(x)
plt.show()
```

As you can see the pie chart draws one piece (called a wedge)

By default the plotting of the first wedge starts from the x-axis and move counterclockwise:


## Labels

Add labels to the pie chart with the **label** parameter.

The **label** parameter must be an array with one label for each wedge:

```python
x=np.array([30,27,40,15])
lab=['bikes','cars','cycles','bus']

plt.pie(x,labels=lab)
plt.show()
```

## Start angle

As mentioned the default start angle is at the x-axis, but you can change the start angle by specifying a **startangle** parameter.

The **startangle** parameter is defined with an angle in degrees, default angle is 0:

```python
x=np.array([30,27,40,15])
lab=['bikes','cars','cycles','bus']

plt.pie(x,labels=lab,startangle=90)
plt.show()
```

## Explode

The **explode** parameter, if specified, and not None, must be an array with one value for each wedge.

Each value represents how far from the center each wedge is displayed:


```python
x=np.array([30,27,40,15])
lab=['bikes','cars','cycles','bus']
ex=[0.2,0,0,0.1]

plt.pie(x,labels=lab,explode=ex)
plt.show()
```

## shadow

Add a shadow to the pie chart by setting the **shadows parameter to True**:

```python
x=np.array([30,27,40,15])
lab=['bikes','cars','cycles','bus']
ex=[0.3,0,0,0]

plt.pie(x,labels=lab,explode=ex,shadow=True)
plt.show()
```

## Colors

You can set the color of each wedge with the **colors** parameter.

The **colors** parameter, if specified, must be an array with one value for each wedge:

```python
x=np.array([25,36,43,14])
lab=['fruits','vegitable','meat','fish']
col=['red','green','brown','yellow']

plt.pie(x,labels=lab,colors=col)
plt.show()
```

## Legend

To add a list of explanation for each wedge, use the **legend()** function:

```python
x=np.array([30,27,40,15])
lab=['bikes','cars','cycles','bus']

plt.pie(x,labels=lab)
plt.legend()
plt.show()
```

## Legend with header

To add a header to the legend, add the **title parameter to the legend function**

```python
x=np.array([30,27,40,15])
lab=['bikes','cars','cycles','bus']

plt.pie(x,labels=lab)
plt.legend(title='vehicles:')
plt.show()
```

# Waffle Chart

A waffle chart is basically a square display, usually consisting of 100 smaller squares arranged in a 10-by-10 layout.

The squares are colored according to the proportions you are aiming to visualize

similarly to how you would color different slices of a pie chart. As this square layout somewhat resembles a delicious waffle


```python
!pip install pywaffle
```

```python

import matplotlib.pyplot as plt
from pywaffle import Waffle

fig=plt.figure(FigureClass=Waffle,rows=5,columns=10,values=[20,50,30,90],figsize=(5,3))
plt.show()
```

```python
data={'apple':40,'cherry':20,'banana':15}
fig=plt.figure(FigureClass=Waffle,rows=5,columns=10,values=data,legend={'loc':'upper left','bbox_to_anchor':(1.1,1)})
plt.show()
```

Title, Legend, Colors, Background Color, Block Color, Direction and Style

```python
data={'apple':40,'cherry':20,'banana':15}

fig=plt.figure(FigureClass=Waffle,rows=5,values=data,colors=['red','black','yellow'],
               title={'label':'fruits sale per day','loc':'left'},labels=[f"{k} ({v}%)" for k, v in data.items()],
               legend={'loc':'lower left','bbox_to_anchor':(0,-0.4),'ncol':len(data),'framealpha':0.5},
              block_arranging_style='snake')

fig.set_facecolor('pink')   # bckgrd color
plt.show()
```

'snake': Fills the blocks in a snake-like pattern.
'new-next': Fills each row or column from a new start point.
'top-left': Fills from top-left, then right, then down.
'bottom-left': Fills from bottom-left, then right, then up.
'top-right': Fills from top-right, then left, then down.
'bottom-right': Fills from bottom-right, then left, then up.

Plot with Icons - Pictogram Chart

```python
data={'apple':40,'cherry':20,'banana':15}
fig=plt.figure(FigureClass=Waffle,rows=5,values=data,legend={'loc':'upper left','bbox_to_anchor':(1.1,1)},
               icons='child',font_size=15,icon_legend=True)
plt.show()
```

# Word Cloud

Word Cloud is a data visualization technique used for representing text data in which the size of each word indicates its frequency or importance.

Significant textual data points can be highlighted using a word cloud.


```python
!pip install wordcloud
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Create a list of word
txt=('hello hello what a funny funny in the morning situation situation i feel so good good to something want go public')

# Create the wordcloud object
wordcloud=WordCloud(width=500,height=500,margin=0).generate(txt)

# Display the generated image:
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```
